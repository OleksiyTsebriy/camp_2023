# Data Science Overview 

## Data science
<font size= 4 style="line-height: 1.5em">
<br>
<b>Data science</b> is a practical application of machine learning with a complete focus on solving real-world problems.
</font> 

<!-- data_science.jpeg -->



<br><br><br><br><br><br>

---
<div align= "right">
5
</div>


<font size= 5 style="line-height: 1.5em">
<b>Data science</b> could be any extraction of relevant insights from data
</font> 

- [data storage](https://www.toptal.com/data-science/data-engineering-guide-to-storages) (part of data science pipeline related to decisions of how/where the data can be used and analyzed.)
- [data lake](https://www.altexsoft.com/blog/datascience/what-is-data-engineering-explaining-data-pipeline-data-warehouse-and-data-engineer-role/) is a vast pool for saving data in its native, unprocessed form  
- [data warehouse](https://en.wikipedia.org/wiki/Data_warehouse) (database system which is designed for analytical instead of transactional work / is a method of centralizing data from different sources into one common repository)
- [data mining](https://en.wikipedia.org/wiki/Data_mining) is looking for hidden, valid, and potentially useful patterns in huge data sets
- [data cleaning](https://en.wikipedia.org/wiki/Data_cleansing) is the process of detecting, correcting or removing corrupt or inaccurate records
- [data validation](https://en.wikipedia.org/wiki/Data_validation) is the process of ensuring data has undergone data cleansing to ensure data is correct and useful
- [data transformation](https://en.wikipedia.org/wiki/Data_transformation) is process of converting data from one format or structure into another format or structure
- [data management](https://en.wikipedia.org/wiki/Data_management) is the practice of organizing and maintaining data processes to meet ongoing information lifecycle needs
- [data analysis](https://en.wikipedia.org/wiki/Data_analysis) is process of inspecting, cleansing, transforming, and modeling data with the goal of discovering useful information, informing conclusions, and supporting decision-making
- [data integration](https://en.wikipedia.org/wiki/Data_integration) is process of combining data residing in different sources and preparing it in unified view
- [data engineering](https://quanthub.com/what-is-data-engineering/) is process of designing and building pipelines that transform and transport data into a format in usable state for analysis and machine learning 
- [data visualization](https://en.wikipedia.org/wiki/Data_visualization) is the graphic representation of data particularly efficient for communicating when the data is numerous
- [data presentation/BI](https://en.wikipedia.org/wiki/Business_intelligence) is "a set of methodologies, processes, architectures, and technologies that transform raw data into meaningful and useful information used to enable more effective strategic, tactical, and operational insights and decision-making
- [data-driven decisions](https://www.tableau.com/learn/articles/data-driven-decision-making) is using facts, metrics, and data to guide strategic business decisions)
- big data 
- artificial intelligence 
- machine learning

# Machine learning

<img src = "img/RL3.jpeg" height=250 width= 150 align="left">

<font size= 4 style="line-height: 1.5em">
<br>
&nbsp; &nbsp;  <b>Machine learning</b> is an field of AI that provides computer systems <br>&nbsp; &nbsp; the ability to automatically learn and improve from experience <br>&nbsp; &nbsp; 
without being explicitly programmed.
<br><br>&nbsp; &nbsp;&nbsp; &nbsp;&nbsp; &nbsp;&nbsp; &nbsp;&nbsp; &nbsp;     Arthur Samuel (1959)

</font>



<img src = "img/RL1.jpeg" height=300 width= 300 align="left">

<font size= 4 style="line-height: 1.5em">
<br>
&nbsp; &nbsp; "A computer program is said to learn from experience E 
    <br>&nbsp; &nbsp;with respect to some class of tasks T<br>&nbsp; &nbsp; and performance measure P 
    <br>&nbsp; &nbsp; if its performance at tasks in T, <br>&nbsp; &nbsp;
as measured by P,<br>&nbsp; &nbsp; improves with experience E”
<br><br>&nbsp; &nbsp;&nbsp; &nbsp;&nbsp; &nbsp;&nbsp; &nbsp;&nbsp; &nbsp;     Tom M. Mitchell (1997)

</font>
<div style="clear:left;"></div>  
<b>Chess playing</b> <br>
T - play chess <br>
E - number of played games <br>
P - probability to win  
<br>
<b>Spam Classification</b>
<br>
T - classify spam  <br>
E - number of labeled data,  <br>
P - accuracy

## Other ML Types 

<img src = "img/RL2.png" height=400 width= 400 align="right">
<font size= 4 style="line-height: 1.5em">
<br> <br>    
- <b>Semi-Supervised Learning Method.</b> Labelling of data is manual work and is very costly as data is huge. Thus, the model first trains under unsupervised learning. This ensures that most of the unlabeled data divide into clusters. Then, the remaining unlabeled data, gets generation of labels.
<br> <br>    
- <b>Self-Supervised Learning,</b> where labelling of the data is not done by humans. Model itself extracts and labels the data with the help of the embedded metadata as supervisory data. 
<br><br> 
- <b>Active Learning,</b> which selects the data points based on certain instances to exclude unwanted data e.g. instead of considering all patients analyze after checking, in an active method, it selects heart ailments in patients above 50 years of age who eat food with high cholesterol.
<br><br> 
- <b>Inductive Learning, Deductive Learning, Transductive Learning</b> - these all about creation of a generalized rule for all the data given to the algorithm.

More details:
- [Types of Machine Learning](https://techvidvan.com/tutorials/types-of-machine-learning/)
- [Self-Supervised Learning](https://perfectial.com/blog/self-supervised-learning/) 
    
</font>
