# Python

Python is an `interpreted`, high-level, general-purpose programming language, created by Guido van Rossum and first released in 1991.

Python 2 language was officially discontinued in 2020.
`Python 3` was released on 2008 is still actual and supported.

You may have **several python interpreters on the same machine**.

There are following approaches of installing python:
- python interpreter on local machine 
- python interpreter of virtual environment on local machine
- (python interpreter of conda or conda virtual environment on local machine)
- python interpreter of docker container 
- Google colab
- etc.

### Launch Jupyter notebook on docker container

1) Install [docker](https://docs.docker.com/get-docker/)<br>
2) check docker ios running:  `docker -v` <br>
3) run command `docker run -it -p 8820:8888 -v "$(pwd):/work" oleksiytsebriy/data_science_env:v0` <br>
**Note**: You will need to wait minute or two while it loads the image first time <br>
4) open in browser `http://localhost:8820/tree`
